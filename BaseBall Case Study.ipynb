```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def vifcheck(x_scaled):
    vif=pd.DataFrame()
    vif['Features']=x_scaled.columns
    vif['variance_inflation_factor']=[variance_inflation_factor(x_scaled.values,i) for i in range (x_scaled.shape[1])]
    print(vif)
```
