# Regression Type 

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
maxRs=0
max_Acc=0

for i in range(1,300):
    x_train,x_test,y_train,y_test=train_test_split(x_reg,y_reg,test_size=.30,random_state=i)
    rf=RandomForestRegressor()
    rf.fit(x_train,y_train)
    pred=rf.predict(x_test)
    acc=r2_score(y_test,pred)
    error=np.sqrt(mean_squared_error(y_test,pred))
    if acc>max_Acc:
        maxRs=i
        max_Acc=acc
        
print(f'Best score{max_Acc} and least error{error} in randomstate{maxRs}')
```

```python
def modelcheck(models,x,y):
    x1_train,x1_test,y1_train,y1_test=train_test_split(x,y,test_size=.30,random_state=rs)
        
    for model in models:
          
            print(f'Model:{model.__class__.__name__} with random_state={rs}')
            model.fit(x1_train,y1_train)
            pred_model=model.predict(x1_test)
            accuracy=r2_score(y1_test,pred_model)
            print(f'Accuracy:{accuracy}')
            mse=mean_squared_error(y1_test,pred_model)
            print('Meansquare_error',mse)
            print('RootMeansquare:',np.sqrt(mse))
            model_cross_score=cross_val_score(model,x,y,cv=5)
            print(model_cross_score)
            print(model_cross_score.mean())
        
            difference= accuracy-model_cross_score.mean()
            print(f'Difference Between Accuracy and Cross_val_score mean: {difference}\n')
        
lin=LinearRegression()
rf1=RandomForestRegressor()
dtc1=DecisionTreeRegressor()
knn1=KNeighborsRegressor()
et1=ExtraTreesRegressor()
ada1=AdaBoostRegressor()
gb1=GradientBoostingRegressor()
for j in [0.1,0.01,0.001]:
    ls1=Lasso(alpha=j)
for k in [0.1,0.01,0.001]:
    rd1=Ridge(alpha=k)
x=x_reg
y=y_reg
rs=202
models=[lin,rf1,dtc1,knn1,et1,ada1,gb1,ls1,rd1]
modelcheck(models,x,y)
```

# Classification Type

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.model_selection import train_test_split
maxAcc=0
maxRs=0
for i in range(1,300):
    x_train,x_test,y_train,y_test=train_test_split(x_class,y_class,test_size=.30,random_state=i)
    rf=RandomForestClassifier()
    rf.fit(x_train,y_train)
    pred=rf.predict(x_test)
    accuracy=accuracy_score(y_test,pred)
    if accuracy>maxAcc:
        maxAcc=accuracy
        maxRs=i
print(f'Best Accuracy for the sampled date is {maxAcc} and randomstate is{maxRs}')
```

```python
def modelcheck(models,x,y):
    x1_train,x1_test,y1_train,y1_test=train_test_split(x,y,test_size=.30,random_state=rs)
        
    for model in models:
          
            print(f'Model:{model.__class__.__name__} with random_state={rs}')
            model.fit(x1_train,y1_train)
            pred_model=model.predict(x1_test)
            accuracy=accuracy_score(y1_test,pred_model)
            print(f'Accuracy:{accuracy}')
            print('Confusion_matrix:',confusion_matrix(y1_test,pred_model))
            print('Classification_report:',classification_report(y1_test,pred_model))
            model_cross_score=cross_val_score(model,x,y,cv=5)
            print(model_cross_score)
            print(model_cross_score.mean())
        
            difference= accuracy-model_cross_score.mean()
            print(f'Difference Between Accuracy and Cross_val_score mean: {difference}\n')
        
rf1=RandomForestClassifier()
dtc1=DecisionTreeClassifier()
knn1=KNeighborsClassifier()
et1=ExtraTreesClassifier()
ada1=AdaBoostClassifier(algorithm='SAMME')
gb1=GradientBoostingClassifier()
x=x_class
y=y_class
rs=50
models=[rf1,dtc1,knn1,et1,ada1,gb1]
modelcheck(models,x,y)
```
